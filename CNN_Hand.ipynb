```python
def output_dim(img: np.array, kernels: np.array, padding: int, strides: int):
    
    channel,img_weight,img_height = img.shape
    out_height = int((img_height+padding-kernels.shape[1])/strides+1)
    out_weight = int((img_weight+padding-kernels.shape[2])/strides+1)
    if out_height%1 != 0 or out_weight%1 !=0:
        print('error')
    out_array = np.zeros(shape = (n_kernels,out_weight,out_height))
    
        
    for x in range(n_kernels):#loop for all out put channel
        for y in range(out_weight):#loop for one img
            for z in range(out_height):
                    result = 0#loop for every convolution
                    for i in range(k_dim):
                        for j in range(k_dim):
                            for color_channel in range(channel):
                                result = result + img[color_channel][i+y*strides][j+z*strides] * kernels[x][i][j][color_channel]
                
                    out_array[x][y][z] = result
    return out_array
    # your code here
    
```

```python
def maxPool2D(array,poolsize,strides):
    height,weight = array.shape
    out_height = int((height-poolsize)/strides) + 1
    out_weight = int((weight-poolsize)/strides) + 1
   
    out_array = np.zeros((out_height,out_weight))
    for i in range(0,out_height):#for each pixel
        for j in range(0,out_weight):
            pool_value = array[i*strides][j*strides]
            for x in range(poolsize):
                for y in range(poolsize):
                    pool_value = max(pool_value,array[i*strides+x][j*strides+y])
                    
            out_array[i][j]=pool_value
    return out_array
            
            
    
```
